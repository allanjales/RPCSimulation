```python
import uproot
import numpy as np
import mplhep as hep
import matplotlib as plt
```

```python
file = uproot.open("histograms.root")
file.keys()
```

```python
hep.style.use("CMS")

fig, ax = plt.pyplot.subplots()
ax.set_xlabel(r"cos$\theta$")
ax.set_ylabel("Events")

hep.cms.text("RPC Simulation")
hep.histplot(file["ElectronCosThetaDistribution"].to_numpy(), ax=ax)
```
