# Definitions

```python
import uproot
import numpy as np
import mplhep as hep
import matplotlib as plt
```

```python
def ShowStats():
    # compute mean and std:
    values, edges = bins
    centers = (edges[:-1] + edges[1:]) / 2
    values = file["CosTheta"].to_numpy()[0]
    
    mean = np.sum(values*centers)/np.sum(values)
    std = np.sqrt(np.sum(values*((centers - mean)**2))/np.sum(values))
    
    annotation = plt.offsetbox.TextArea(f"Total: {np.sum(values):.2e}" \
                        + "\n" + f"Mean: {mean:.2e}" \
                        + "\n" + f"Std: {std:.2e}", textprops=dict(size=14))
    
    at = plt.offsetbox.AnchoredOffsetbox('upper right', child=annotation)
    at.patch.set_facecolor('None')
    ax.add_artist(at)
```
